```python
# clase factura,atributos folio, cliente y total
# funciones leerdatos(), mostrardatos()
# crear 2 objetos factura1, factura2

class Factura:
    def __init__(self, folio, cliente, total):
        self.folio = folio
        self.cliente = cliente
        self.total = total

    def leerdatos(self):
        self.folio = input("Ingrese el folio: ")
        self.cliente = input("Ingrese el cliente: ")
        self.total = input("Ingrese el total: ")
        return self.folio, self.cliente, self.total

    def mostrardatos(self):
        print("Folio: ", self.folio)
        print("Cliente: ", self.cliente)
        print("Total: ", self.total)
        return self.folio, self.cliente, self.total
# fin de la clase Factura


print("factura numero 1")
factura1 = Factura("", "", "") # creando el objeto factura1
factura1.leerdatos()
factura1.mostrardatos()
print("*_*_*_*_*_*_+_*_*_*_*_*_*")
print("factura numero 2")
factura2 = Factura("", "", "") # creando el objeto factura2
factura2.leerdatos()
factura2.mostrardatos()

print("software desarrollado por Hugo Fabian NL = 15 + 3 = 18")
```
